## Strings 1

Write a function that returns a string that states the number of users logged into the system. The function should print the number in single quotes and correctly handle single and no users being logged in.
E.g.
- There are '5' users logged into the system.
- There is 1 user logged into the system.
- There are no users logged into the system.

````
def pretty_print_user_count(count: int) -> str:
  if count > 1:
    return f"There are '{count}' users logged into the system."
  if count == 1:
    return f"There is '{count}' user logged into the system."
  return f"There are no users logged into the system."
````

````
check(pretty_print_user_count(8), "There are '8' users logged into the system.")
check(pretty_print_user_count(1), "There is '1' user logged into the system.")
check(pretty_print_user_count(0), "There are no users logged into the system.")
check(pretty_print_user_count(-1), "There are no users logged into the system.")
````

## Strings 2

Complete the following function that takes a string and returns a string consisting of the first 10 characters of the string followed by a space and then the last 6 characters of the string. If the input string is less than 16 characters, return an empty string.

````
def extract_timings(string: str) -> str:
  if len(string) >= 16:
    return f'{string[:10]} {string[-6:]}'
  return ''
````

````
check(extract_timings('2019-06-25. EmployeeController.get 1203ms'), '2019-06-25 1203ms')
check(extract_timings('2019-06-26. JobController.get 0130ms'), '2019-06-26 0130ms')
check(extract_timings('Server Shutdown'), '')
````

## Strings 3

Write a function that accepts a string and converts camel case to snake case. Strings starting with a capital letter may be ignored.

E..g.

doSomething becomes do_something

addElement becomes add_element

````
def camel_case(s):
  solution = ''
  for c in s:
    if c.islower():
      solution += c
    else:
      solution += f'_{c.lower()}'
  return solution
````

````
check(camel_case('doSomething'), 'do_something')
check(camel_case('addElement'), 'add_element')
check(camel_case('test'), 'test')
````

## Strings 4: Not Bad (Optional)

Write a function that takes a string and finds the first occurrences of 'not' and 'bad'. If 'bad' follows 'not' then return a new string where the entire substring between 'not' and 'bad' (inclusive) is replaced with 'good'.

E.g.

'This weather is not bad' become 'This weather is good'.

````
def not_bad(s):
  n = s.find('not')
  b = s.find('bad')
  if n != -1 and b != -1 and b > n:
    s = s[:n] + 'good' + s[b+3:]
  return s
````

````
check(not_bad('This weather is not bad'), 'This weather is good')
check(not_bad('I am not saying the experience was bad'), 'I am good')
check(not_bad('This is a bad idea, let\'s not do it'), 'This is a bad idea, let\'s not do it')
````
